```python
# Sort bookings by price (highest first)
bookings_sorted = bookings.sort_values('Price', ascending=False)

# Rank customers by total bookings
bookings['Customer Rank'] = bookings.groupby('Customer Name')['Booking ID'].transform('count')
```

```python
#Identify missing data in each column
missing_values = bookings.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
#Handling Missing Data
bookings["Customer Name"].fillna("Unknown", inplace=True)
bookings["Status"].fillna(bookings["Status"].mode()[0], inplace=True)
```

```python
bookings["Booking Date"] = pd.to_datetime(bookings["Booking Date"])
bookings["Booking Date"].fillna(bookings["Booking Date"].mode()[0], inplace=True)
```

```python
bookings["Price"].fillna(bookings["Price"].median(), inplace=True)
bookings["Duration (mins)"].fillna(bookings["Duration (mins)"].median(), inplace=True)
```

```python
bookings["Customer Email"].fillna("Not Provided", inplace=True)
bookings["Customer Phone"].fillna("Not Provided", inplace=True)
```

```python
#Check for duplicate rows
print("Duplicate Rows: ", bookings.duplicated().sum())
bookings = bookings.drop_duplicates()
```

```python
#Ensure each column has the right data type
bookings["Booking Date"] = pd.to_datetime(bookings["Booking Date"])  # Convert dates
bookings["Price"] = pd.to_numeric(bookings["Price"], errors='coerce')  # Convert price to numeric
bookings["Duration (mins)"] = pd.to_numeric(bookings["Duration (mins)"], errors='coerce')  # Convert duration to numeric
```

```python
Q1 = bookings["Price"].quantile(0.25)
Q3 = bookings["Price"].quantile(0.75)
IQR = Q3 - Q1
bookings = bookings[(bookings["Price"] >= (Q1 - 1.5 * IQR)) & (bookings["Price"] <= (Q3 + 1.5 * IQR))]

Q1 = bookings["Duration (mins)"].quantile(0.25)
Q3 = bookings["Duration (mins)"].quantile(0.75)
IQR = Q3 - Q1
bookings= bookings[(bookings["Duration (mins)"] >= (Q1 - 1.5 * IQR)) & (bookings["Duration (mins)"] <= (Q3 + 1.5 * IQR))]
```

```python
bookings["Customer Name"] = bookings["Customer Name"].str.strip().str.title()
bookings["Booking Type"] = bookings["Booking Type"].str.strip().str.title()
bookings["Status"] = bookings["Status"].str.strip().str.capitalize()
```

```python
bookings["Status"].replace({"cancelled": "Cancelled", "confirmed ": "Confirmed"}, inplace=True)
```

```python
bookings["Valid Email"] = bookings["Customer Email"].str.contains(r"[^@]+@[^@]+\.[^@]+", na=False)
```

```python
bookings["Customer Phone"] = bookings["Customer Phone"].astype(str).str.replace(r"\D", "", regex=True)  # Keep only digits
```

```python
# Pivot table
pivot = bookings.pivot_table(index='Booking Type', columns='Class Type', values='Price', aggfunc='sum')

print(pivot)
```

```python
# total bookings
total_bookings = bookings.shape[0]  # Number of bookings
total_revenue = bookings["Price"].sum()  # Total revenue

print(f"Total Bookings: {total_bookings}")
print(f"Total Revenue: ${total_revenue:.2f}")

```

```python
# Count of each booking type
booking_type_counts = bookings["Booking Type"].value_counts()

# Plot the distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=booking_type_counts.index, y=booking_type_counts.values, palette="viridis")
plt.title("Number of Bookings by Type")
plt.xlabel("Booking Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
booking_type_counts
```

```python
#monthly bookings
# Convert booking_date to datetime
bookings["Booking Date"] = pd.to_datetime(bookings["Booking Date"])

# Extract year and month
bookings["Year_Month"] = bookings["Booking Date"].dt.to_period("M")

# Group by month
monthly_trend = bookings.groupby("Year_Month")["Price"].sum()
 
# Plot revenue trend
plt.figure(figsize=(10, 5))
monthly_trend.plot(marker='o', color="b", linestyle='-')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue ($)")
plt.xticks(rotation=45)
plt.grid()
plt.show()
monthly_trend
```

```python
#top 5 booking customer names
top_customers = bookings.groupby("Customer Name")["Price"].sum().sort_values(ascending=False).head(5)
print("Top 5 Customers by Revenue:\n", top_customers)
```

```python
# Extract weekday
bookings["Weekday"] = bookings["Booking Date"].dt.day_name()

# Count of bookings per day
weekday_counts = bookings["Weekday"].value_counts()

# Plot peak booking days
plt.figure(figsize=(8, 5))
sns.barplot(x=weekday_counts.index, y=weekday_counts.values, palette="coolwarm")
plt.title("Bookings by Day of the Week")
plt.xlabel("Weekday")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.show()
weekday_counts
```

```python
# Group revenue by booking type
revenue_by_type = bookings.groupby("Booking Type")["Price"].sum()

# Pie Chart for revenue distribution
plt.figure(figsize=(7, 7))
revenue_by_type.plot.pie(autopct="%1.1f%%", cmap="Set3", startangle=90)
plt.title("Revenue Contribution by Booking Type")
plt.ylabel("")  # Hide y-label
plt.show()
revenue_by_type
```
